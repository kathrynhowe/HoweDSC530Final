Since diagnosis is a yes or no, I decided to introduce the MMSE as a method to compare the diagnosed and undiagnosed patients. Here there is a clear region where those undiagnosed have higher MMSE scores, however that seems uncorrelated to their quality of diet.

```python
class CorrelationPermute(thinkstats2.HypothesisTest):

    def TestStatistic(self, data):
        xs, ys = data
        test_stat = abs(thinkstats2.Corr(xs, ys))
        return test_stat

    def RunModel(self):
        xs, ys = self.data
        xs = np.random.permutation(xs)
        return xs, ys
```

```python
cleaned = diagnosed.dropna(subset=['Age', 'MMSE'])
data = cleaned.Age.values, cleaned.MMSE.values
ht = CorrelationPermute(data)
pvalue = ht.PValue()
pvalue
```

```python
undiagnosed_cleaned = not_diagnosed.dropna(subset=['Age', 'MMSE'])
data = undiagnosed_cleaned.Age.values, undiagnosed_cleaned.MMSE.values
ht = CorrelationPermute(data)
pvalue = ht.PValue()
pvalue
```

I performed the same analysis on both the diagnosed and undiagnosed data frames. In the diagnosed dataset, the returned p value was 0.172. In the undiagnosed patients, the p value returned was 0.277. The values are higher which indicates that there is likely no correlation and any potential correlation is just due to chance.
